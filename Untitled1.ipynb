```python
# ============================
# ATM Face Recognition System - Training Script
# ============================

# STEP 1: Install required packages
!pip install face_recognition opencv-python dlib scikit-learn kaggle

# STEP 2: Upload kaggle.json for dataset download
from google.colab import files
print("🔐 Upload your kaggle.json file:")
uploaded = files.upload()

# STEP 3: Set up Kaggle API key
!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json

# STEP 4: Download and unzip the LFW dataset
!kaggle datasets download -d jessicali9530/lfw-dataset
!unzip -q lfw-dataset.zip
!mv lfw-deepfunneled lfw  # Rename for easier access

# STEP 5: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# STEP 6: Prepare directory to save the model
import os
model_dir = '/content/drive/MyDrive/ATM_SECURITY_MODELS'
os.makedirs(model_dir, exist_ok=True)

# STEP 7: Import libraries
import face_recognition
import pickle
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import cv2
import random

# STEP 8: Define dataset processing function
def process_lfw_dataset(sample_limit=None):
    """
    Process the LFW dataset and extract face encodings

    Args:
        sample_limit (int): Limit the number of people to process (None for all)

    Returns:
        tuple: (encodings, names)
    """
    known_face_encodings = []
    known_face_names = []

    lfw_dir = "lfw"
    people_dirs = sorted(os.listdir(lfw_dir))

    if sample_limit:
        people_dirs = people_dirs[:sample_limit]
        print(f"📦 Limited to first {sample_limit} people")
    else:
        print(f"📦 Processing all {len(people_dirs)} people")

    for person_dir in tqdm(people_dirs):
        person_dir_path = os.path.join(lfw_dir, person_dir)

        if os.path.isdir(person_dir_path):
            image_files = os.listdir(person_dir_path)

            for image_file in image_files[:3]:  # Limit to 3 images/person
                image_path = os.path.join(person_dir_path, image_file)

                try:
                    image = face_recognition.load_image_file(image_path)
                    face_locations = face_recognition.face_locations(image)

                    if face_locations:
                        face_encoding = face_recognition.face_encodings(image, [face_locations[0]])[0]
                        known_face_encodings.append(face_encoding)
                        known_face_names.append(person_dir.replace('_', ' ').title())
                except Exception as e:
                    print(f"⚠️ Error processing {image_path}: {e}")

    print(f"✅ Processed {len(known_face_encodings)} face images from {len(set(known_face_names))} people")
    return known_face_encodings, known_face_names

# STEP 9: Process dataset and save encodings
encodings, names = process_lfw_dataset(sample_limit=500)

if encodings:
    data = {"encodings": encodings, "names": names}

    encodings_file = os.path.join(model_dir, "face_encodings.pickle")
    with open(encodings_file, "wb") as f:
        pickle.dump(data, f)

    print(f"✅ Face encodings saved to Google Drive: {encodings_file}")

    # Also allow download
    with open("face_encodings.pickle", "wb") as f:
        pickle.dump(data, f)

    print("⬇️ Downloading face_encodings.pickle to your computer...")
    files.download("face_encodings.pickle")
else:
    print("❌ No face encodings were generated.")

# STEP 10: Define test function
def test_face_recognition(num_tests=5):
    unique_names = list(set(names))
    test_names = random.sample(unique_names, min(num_tests, len(unique_names)))

    for name in test_names:
        person_dir = name.lower().replace(' ', '_')
        person_dir_path = os.path.join("lfw", person_dir)

        image_files = os.listdir(person_dir_path)
        if len(image_files) > 3:
            test_image_file = image_files[-1]
            test_image_path = os.path.join(person_dir_path, test_image_file)

            image = face_recognition.load_image_file(test_image_path)
            face_locations = face_recognition.face_locations(image)

            if face_locations:
                face_encoding = face_recognition.face_encodings(image, [face_locations[0]])[0]

                matches = face_recognition.compare_faces(encodings, face_encoding)
                face_distances = face_recognition.face_distance(encodings, face_encoding)
                best_match_index = np.argmin(face_distances)

                matched_name = names[best_match_index] if matches[best_match_index] else "Unknown"

                rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                top, right, bottom, left = face_locations[0]
                cv2.rectangle(rgb_image, (left, top), (right, bottom), (0, 255, 0), 2)

                plt.figure(figsize=(8, 6))
                plt.imshow(rgb_image)
                plt.title(f"Actual: {name}\nPredicted: {matched_name}")
                plt.axis('off')
                plt.show()

# STEP 11: Run test on random faces
test_face_recognition(5)
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model, load_model
```

```python
def create_deep_face_model(num_classes):
    """
    Create a deep learning model for face recognition based on ResNet50

    Args:
        num_classes: Number of people to recognize

    Returns:
        Compiled Keras model
    """
    # Start with pretrained ResNet50
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

    # Add custom classification layers
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    predictions = Dense(num_classes, activation='softmax')(x)

    # Create model
    model = Model(inputs=base_model.input, outputs=predictions)

    # Freeze base model layers
    for layer in base_model.layers:
        layer.trainable = False

    # Compile model
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    return model
```
